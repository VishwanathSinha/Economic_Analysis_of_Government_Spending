## Policy and Spending Misalignment Analysis
### ○ Investigate and analyze the alignment between policy declarations, budget allocations, and actual spending.

```python
# Select relevant columns
relevant_columns = ['Project Name', 'Financing Type', 'Current Project Cost','Grant Amount']
policy_budget_data = df2[relevant_columns].copy()
```

```python
policy_budget_data
```

```python
# Group by 'Financing Type' and calculate total budget allocation and actual spending
budget_actual_alignment = policy_budget_data.groupby('Financing Type').agg({'Current Project Cost': 'sum','Grant Amount':'sum'})
budget_actual_alignment.rename(columns={'Current Project Cost': 'Total Spending','Grant Amount':'Budget Allocation'}, inplace=True)
```

```python
budget_actual_alignment
```

```python
#Calculate the variance between budget allocations and actual spending for each financing type. Positive variances indicate overspending, while negative variances indicate underspending.
budget_actual_alignment['Variance'] = budget_actual_alignment['Budget Allocation'] - budget_actual_alignment['Total Spending']
```

```python
budget_actual_alignment['Alignment (%)'] = (budget_actual_alignment['Total Spending'] / budget_actual_alignment['Budget Allocation']) * 100
```

```python
import matplotlib.pyplot as plt

budget_actual_alignment.plot(kind='bar', y=['Budget Allocation', 'Total Spending'], figsize=(10, 6))
plt.xlabel('Financing Type')
plt.ylabel('Amount')
plt.title('Budget Allocation vs. Actual Spending by Financing Type')
plt.legend(["Budget Allocation", "Total Spending"])
plt.show()
```
