## Sector-Wise Analysis
### ○ Evaluate government spending within various sectors to pinpoint instances of budget booms and busts.
### ○ Compare and contrast spending trends across different sectors.

```python
# Group data by Sector 1 and calculate the sum of spending for each sector.
sector1_spending = df2.groupby('Sector 1')['Current Project Cost'].sum()

# You can do the same for Sector 2 and Sector 3.
sector2_spending = df2.groupby('Sector 2')['Current Project Cost'].sum()
sector3_spending = df2.groupby('Sector 3')['Current Project Cost'].sum()


```

```python
aggregated_data = pd.DataFrame({
    'Sector 1 Spending': sector1_spending,
    'Sector 2 Spending': sector2_spending,
    'Sector 3 Spending': sector3_spending
})
```

```python
aggregated_data = aggregated_data.reset_index()
```

```python
aggregated_data.head()
```
